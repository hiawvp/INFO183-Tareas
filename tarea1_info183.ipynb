## Pregunta 2: An√°lisis de Fourier (40%)

- Escriba una funci√≥n que reciba la se√±al y retorne el espectro de amplitud, el espectro de fase y las frecuencias asociadas a la se√±al. 




>
>Sea la transformada de Fourier de una funci√≥n general $s(t)$
>
>Llamamos a $S(\omega)$ el espectro de $s(t)$ 
>
>El espectro es un n√∫mero complejo que podemos escribir en notaci√≥n polar como
>
>$$
S(\omega) = |S(\omega)| e^{j\Phi(\omega)},
$$
>
>donde $|S(\omega)|$ se conoce como **espectro de amplitud** y $\Phi(\omega)$ como **espectro de fase**

```python
import scipy.fft as sfft
def espectros(s, Fs):
    S = sfft.fft(s)
    #espectro de amplitud
    SA = np.absolute(S)        
    #fases
    SP = np.angle(S)
    freqs = sfft.fftfreq(len(S), 1/Fs)
    return [SP, SA, freqs] #espectro de fase, espectro de amplitud y frecuencias asociadas
```

- Muestre dos gr√°ficos con el espectro de amplitud y espectro de fase en funci√≥n de la frecuencia, respectivamente. Comente sobre lo que observa


```python
SP, SA, freqs = *espectros(x, Fs),
fig, ax = plt.subplots(1, 2, figsize=(9,4))
ax[0].grid()
ax[1].grid()

ax[0].set_xlabel("Frecuencia")
ax[1].set_xlabel("Frecuencia")

ax[0].set_xlabel("Amplitud")
ax[1].set_xlabel("Fase")

ax[0].plot(sfft.fftshift(freqs),
           sfft.fftshift(SA) * 2 / n_muestras)
ax[0].set_title('Espectro de amplitud')

ax[1].plot(sfft.fftshift(freqs),
           sfft.fftshift(SP))
ax[1].set_title('Espectro de fase')
plt.show()
```

###### Comentarios

Podemos apreciar claramente en el espectro de amplitud que existen 7 componentes de frecuencia acompa√±ados de un ruido blanco. Adem√°s podemos ver que existen 3 componentes imaginarios y 4 reales.


```python
display(np.max(SP))
display(np.min(SP))
display(np.average(SP))
```

Por otro lado, en el espectro de fase, lo unico que podemos distinguir es la existencia de ruido entre -$\pi$ y $\pi$ 

- Escriba una funci√≥n que reciba un n√∫mero entero $M$ y que retorne un espectro de magnitud modificado donde los $M$ componentes mayores se mantienen y todos los dem√°s se hacen igual a cero, retorne tambi√©n el √≠ndice de los $M$ componentes mayores


```python
def m_mayores(SA, M):
    #tomar los M componentes mayores (indices)
    idxs = np.argsort(SA)[-M:]
    #ndarray de ceros
    modified_SA = np.zeros_like(SA)
    #reemplazar por los M componentes seleccionados donde corresponda
    modified_SA[idxs] = SA[idxs]
    #retornar espectro modificado y los indices
    return [modified_SA, idxs]
```

- Muestre graficamente el espectro de magnitud modificado con $M=7$. Imprima los valores de las **frecuencias, amplitudes y fases** de los $M$ componentes ordenados en orden descedente seg√∫n su amplitud

```python
### amplitud escalada

M = 7
M_SA, idxs = *m_mayores(SA, M),

fig, ax = plt.subplots(figsize=(6,4))
ax.plot(sfft.fftshift(freqs),
        sfft.fftshift(M_SA) * 2 / n_muestras)
ax.set_title(f'Espectro de amplitud modificado\n'
            +f'M = {M} mayores amplitudes')
plt.show()
```

```python
### amplitud escalada

M_comps = np.empty(shape=(M, 3))
#cambiar el orden de los indices a descendente
idxs_decrec = np.flip(idxs)
#freqs
M_comps[:, 0] = freqs[idxs_decrec]
#amplitudes
M_comps[:, 1] = SA[idxs_decrec] * 2 / n_muestras
#fases
M_comps[:, 2] = SP[idxs_decrec]
np.set_printoptions(precision=3, suppress=True)
print(" [ FREQ | AMPLITUD | FASE ]")
print(M_comps)
```

## Pregunta 3: Reconstrucci√≥n (40%)

Use la transformada de Fourier inversa para reconstruir la se√±al en base al espectro de amplitud modificado y el espectro de fase. 

Para $M=[100, 10, 7, 5, 3, 1]$ (puede usar otros valores si lo considera necesario en su an√°lisis) muestre

- Una gr√°fica de linea (`plt.plot`) con las partes real e imaginaria de la se√±al reconstruida en funci√≥n del tiempo
- Una nube de puntos (`plt.scatter`) con la parte imaginaria en funci√≥n de la parte real de la se√±al reconstruida

Describa y comente lo que observa, en particular compare cada resultado con el caso `M=len(x)` (caso original) y discuta sobre lo que ocurre para $M>7$, $M=7$ y $M<7$ con respecto al ruido y la informaci√≥n relevante de la se√±al ¬øPueden ver al elefante en el ruido? 

```python
def graficar_a(axis, REAL, IMAG, M):
    global t
    #mostrar solo una parte de la grafica
    axis.margins(x=-0.4, y=0.)
    
    axis.plot(t, REAL, label='Parte real')
    axis.plot(t, IMAG, label='Parte imaginaria')
    axis.set_xlabel('tiempo')
    axis.set_ylabel('y')
    axis.legend()
    axis.set_title(f"Parte real y parte imaginaria\n en funci√≥n del tiempo")
```

```python
def graficar_b(axis, REAL, IMAG, M):
    axis.scatter(IMAG, REAL)
    axis.set_xlabel('parte real')
    axis.set_ylabel('parte imaginaria')
    axis.set_title("Parte real en funci√≥n\nde la parte imaginaria")
```

```python
try:
    plt.close(fig)
except NameError:
    pass
Ms = [100, 10, 9, 8, 7, 5, 3, 1]
axs = [plt.subplots(1, 2, figsize=(8, 4), tight_layout=True) for m in Ms]
for i, (M, (fig, ax)) in enumerate(zip(Ms, axs)):
    #filtrar
    M_SA = m_mayores(SA, M)[0]
    #reconstruir
    M_s = M_SA * np.exp(1j * SP)
    IS = sfft.ifft(M_s)
    p_real = np.real(IS)
    p_imag = np.imag(IS)
    fig.suptitle(f"Se√±al reconstruida \ncon los {M} mayores componentes.")
    graficar_a(ax[0], p_real, p_imag, M)
    graficar_b(ax[1], p_real, p_imag, M)
plt.show()
    
```

###### COMENTARIOS 
* A medida que disminuimos el numero de componentes que se visualizan podemos ver una imagen mejor formada del elefante. Este efecto lo vemos en los valores de M pertenecientes a [100,7], donde 7 es el valor donde mejor se ve la imagen, debido a que con este valor se elimina el ruido completamente. 


* Para los valores de M entre [10,7], podemos apreciar que, en el grafico de parte imaginaria vs parte real, observamos solo una disminuci√≥n del bordeado de la imagen, esto debido a que, al seleccionar las 7 mayores amplitudes, obtenemos lo que suponemos ser√≠an todas las componentes frecuenciales de la se√±al, y al tomar valores mayores a este no estamos obteniendo informaci√≥n necesaria/fundamental acerca de esta, sino que solo incorporamos leves variaciones a la se√±al real pero que no alcanzan a obstruirla. 


* Si visualizamos menos de 7 componentes, nos enfrentamos a una p√©rdida de informaci√≥n, llegando a un punto (M = 1) en el cual solo podemos apreciar un √∫nico componente que, en la gr√°fica de la izquierda (parte real e imaginaria en funcion del tiempo) posee forma sinusoidal y en el gr√°fico de la derecha (parte imaginaria vs parte real) simplemente vemos una circunferencia.
